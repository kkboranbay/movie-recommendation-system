```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
```

```python
movies = pd.read_csv('movies.csv')
```

```python
ratings = pd.merge(movies, ratings).drop(['genres','timestamp'], axis=1)
```

```python
ratings["title"] = ratings["title"].str.lower()
ratings["title"] = ratings["title"].str.replace(r"\(.*\)","")
```

```python
user_ratings = ratings.pivot_table(index=['title'],columns=['userId'],values='rating')
```

```python
user_ratings
```

```python
user_ratings = user_ratings.dropna(thresh=10,axis=0).fillna(0)
```

```python
user_ratings
```

```python
similarities_df = user_ratings.T.corr(method='pearson')
```

```python
similarities_df
```

```python
import joblib
```

```python
joblib.dump(similarities_df, 'similarities.pkl')
```

```python
moviesData=ratings.drop(['userId','rating'],axis=1)
moviesData=moviesData.drop_duplicates()
moviesData=moviesData.sort_values("movieId")
moviesData=moviesData.reset_index(drop=True)
moviesData["title"] = moviesData["title"].str.lower() 
moviesData["title"] = moviesData["title"].str.replace(r"\(.*\)","")
moviesData.head()
```

```python
joblib.dump(moviesData, 'moviesData.pkl')
```
